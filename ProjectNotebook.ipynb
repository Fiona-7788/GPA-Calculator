# GPA Calculator

This project is a GPA Calculator built in Python. It allows a student
to track their courses and letter grades, compute a weighted cumulative
GPA, simulate how a future course might affect their GPA, visualize
their grades as a bar chart, and save/load their data to a file so it
persists between sessions.

The project is built around a `Student` class in `gpa_module.py`. Each
`Student` object stores a list of courses and provides methods for adding
courses, computing GPA, running what-if simulations, plotting grades,
and saving/loading data via JSON.

## Project Code

Run the cells below to see the GPA Calculator in action.

```python
from gpa_module import Student, letter_to_points
```

```python
student = Student("Alex")
student.add_course("COGS 18", "A", 4)
student.add_course("MATH 20A", "B+", 4)
student.add_course("WCWP 10A", "A-", 4)
student.add_course("COGS 9", "B+", 4)

student.display_transcript()
print(f"What if I get an A in a 4-unit course? GPA: {student.what_if('A', 4)}")
print(f"What if I get a C in a 4-unit course? GPA: {student.what_if('C', 4)}")
student.plot_grades()
student.save("alex.json")
```
